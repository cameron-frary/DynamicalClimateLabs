#### Finding $\epsilon$

1. **Write a Python function that solves for OLR in our two-layer atmosphere model with `Ts = 288`, `T0 = 275`, `T1 = 230`.**
2. **Plot OLR for a range of $\epsilon$ from 0 to 1. You will need the function `linspace` from `numpy`.**
3. **Present-day OLR is about 238.5 W/m$^2$. Find the emissivity value that roughly corresponds to this OLR.**


```python
import numpy as np
import matplotlib.pyplot as plt

sig = 5.67e-8 # W/m^2/K^4 Stefan-Boltzmann constant 

Ts = 288  # K; Surface temp
T0 = 275  # K; Layer 0 temp
T1 = 230  # K; Layer 1 temp

# Stefan-Bolzmann: F = e*sig*T^4

def calc_OLR(Ts, T0, T1, e):
    return sig*((1-e)**2*Ts**4 + e*(1-e)*T0**4 + e*T1**4)

e_list = np.linspace(0, 1, 101)  # List of emissivity values from 0 to 1 in intervals of 0.01

plt.title("OLR vs $\epsilon$")
plt.plot(e_list, calc_OLR(Ts, T0, T1, e_list), label = "Calculated OLR")  # Plot OLR vs emissivity
plt.axhline(y = 238.5, linestyle = ":", label = "Current OLR (W/m$^2$)")
plt.grid()
plt.xlabel("$\epsilon$")
plt.ylabel("OLR (W/m$^2$)")
plt.legend()
plt.show()

plt.title("OLR vs $\epsilon$, zoomed")
plt.plot(e_list, calc_OLR(Ts, T0, T1, e_list), label = "Calculated OLR")  # Plot OLR vs emissivity
plt.axhline(y = 238.5, linestyle = ":", label = "Current OLR (W/m$^2$)")
plt.grid()
plt.xlim(0.55, 0.6)
plt.ylim(225, 250)
plt.xlabel("$\epsilon$")
plt.ylabel("OLR (W/m$^2$)")
plt.legend()
plt.show()

# e = 0.586

```

**We can also solve the two-layer OLR equation algebraically (using the quadratic formula) and get $\epsilon$ = 0.586 and $\epsilon$ = 3.93. Which one should we use?**

Because $0 \leq \epsilon \leq 1$ in all circumstances, we should use $\epsilon = 0.586$.

```python
eps = 0.586  # We use only realistic value for epsilon
Ts = 288
T0 = 275
T1 = 230
```

#### Contributions to OLR from each layer

**Calculate the contributions to OLR from each layer (surface, layer 0, layer 1). These should sum to the observed OLR (238.5 W/m2).**

```python
OLR_s = (1-eps)**2 * sig * Ts**4  # W/m^2; OLR from surface passing through two atmos layers
OLR_0 = eps * (1-eps) * sig * T0**4  # W/m^2; OLR from L0 passing through one atmos layer
OLR_1 = eps * sig * T1**4  # W/m^2; OLR from L1 emitting straight into space

print(OLR_s)
print(OLR_0)
print(OLR_1)

print(OLR_s + OLR_0 + OLR_1)
```

#### OLR response to changing greenhouse gas concentrations

##### What happens to radiative forcing at each layer with a 2% increase in e? What is the total effect on radiative forcing?

```python
delta_eps = eps * 0.02
```

```python
#  Component originating from the surface
Rs = -1*delta_eps*-2*(1-eps)*sig*Ts**4
Rs
```

```python
#  Component originating from level 0
R0 = -1*delta_eps*(1-2*eps)*sig*T0**4
R0
```

```python
#  Component originating from level 1
R1 = -1*delta_eps*sig*T1**4
R1
```

```python
R = Rs + R0 + R1
R
```

```python
print(round(R, 2))
```

**Calculate Earth's equilibrium temperature with S0 and with S0 + R. How much warming does our 2% change in emissivity cause?**


```python
A = 0.3
S = 1370
gh_eff = 33

def calc_T_equil(A, S, sig, gh_eff, R=0): # define the function: func_name(input1, input2)
    shortwave_comp = (S/4)*(1-A)  # W/m^2; shortwave flux not reflected
    longwave_comp = R  # W/m^2; radiative forcing, flux from longwave trapped by increased eps
    return np.power((shortwave_comp + longwave_comp)/sig, (1/4)) + gh_eff - 273  # C; Stefan-Boltzmann

SO_equil_temp = calc_T_equil(A, S, sig, gh_eff)  # C; equil temp without radiative forcing
SO_R_equil_temp = calc_T_equil(A, S, sig, gh_eff, R)  # C; equil temp with radiative forcing

print(f"Without radiative forcing: {round(SO_equil_temp,2)}\N{DEGREE SIGN}C")
print(f"With radiative forcing: {round(SO_R_equil_temp,2)}\N{DEGREE SIGN}C")
```

**The current concentration of CO$_2$ in the atmosphere is about 420 ppm. Pre-industrial CO$_2$ was about 280 ppm. Earth has warmed about 1.2 degrees C during this time. Think about the relationship between CO$_2$ and $\epsilon$ and write your thoughts below.**

As CO$_2$ concentrations increase, the atmosphere absorbs more longwave radiation which results in an increased $\epsilon$. This warms the Earth.
